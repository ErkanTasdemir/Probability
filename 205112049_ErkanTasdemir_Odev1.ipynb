Soru : Rastgele iki noktasından kırılan bir çubuğun parçalarının üçgen oluşturma olasılığı nedir?

Çözüm : Üçgen olabilme kuralı her hangi iki kenarın toplamının diğer kenardan büyük olma durumudur. Programın yapması gerekense defalarca iki rastgele noktadan kırıp oluşan parçaları karşılaştırmaktır.

Aşağıdaki kod, 10000 deneme sonucunda oluşan parçaları kurala göre karşılaştırıp üçgen oluşturabilme olasılığını hesaplamaktadır.

```python
import random
ucgen=0
for i in range(10000):
  nokta = [random.random(),random.random()]
  nokta.sort()
  a,b,c = nokta[0],nokta[1]-nokta[0],1-nokta[1]
  if a+b>c and a+c>b and b+c>a:
    ucgen+=1
print(ucgen/10000)
```
