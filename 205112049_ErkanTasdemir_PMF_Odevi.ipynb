**Probability Mass Function(PMF) - Olasılık Kütle Fonksiyonu**

 **Problem :** Radyoaktif bir kayanktan salınan fotonların verdiği ortalama sayım 20 dakikada 5 dir. 
İkinci ölçümde 10 dakikada 4 sayım alma olasılığı nedir?

**Çözüm :**

Poisson dağılımı bir zaman aralığı içinde bir olayın belirli sayıda olma olasılığı ile ilgilenen sonlu(discrete) bir olasılık dağılımıdır. 

image.png

λ = 20 dakikada 5 sayım => 10 dakikada 2,5 sayım

burada x = 4 sayım


P(X=4)= (λ^4*e^-λ)/4! = (2.5^4*e^-2.5)/4! = 0,1336


```python

#Importing the required modules
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson
 
#Applying the poisson class methods
x = np.arange(0,10)
pmf = poisson.pmf(x,2.5)
 
#Visualizing the results
sb.set_style('whitegrid')
plt.vlines(x ,0, pmf, colors='k', linestyles='-', lw=6)
plt.ylabel('Probability')
plt.xlabel('SÜRE')
plt.show()
```
